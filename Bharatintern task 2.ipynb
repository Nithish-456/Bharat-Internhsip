```python
x=df.drop(['species'],axis=1)
y=df['species']
```

```python
from sklearn.model_selection import train_test_split
xtr,xte,ytr,yte=train_test_split(x,y,test_size=0.4,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
dt=DecisionTreeClassifier(random_state=0)
dt.fit(xtr,ytr)
```

# Feature Importance

```python
df1=pd.DataFrame({
    "Feature Name": xtr.columns,
    "Importance":dt.feature_importances_
})
f1=df1.sort_values(by="Importance",ascending=False)
f2=f1.head(10)
plt.figure(figsize=(8,6))
sns.barplot(data=f2,x="Importance",y="Feature Name")
plt.title("Features impact on the model")
plt.xlabel('Importance')
plt.ylabel('Feature Name')
plt.show()
```

```python
yp1=dt.predict(xte)
yp1
```

```python
gb=GaussianNB()
gb.fit(xtr,ytr)
```

```python
yp2=gb.predict(xte)
yp2
```

```python
kn=KNeighborsClassifier(n_neighbors=5,p=2)
kn.fit(xtr,ytr)
```

```python
yp3=kn.predict(xte)
yp3
```

```python
from sklearn.metrics import accuracy_score,f1_score,confusion_matrix
a=accuracy_score(yte,yp1)
b=f1_score(yte,yp1,average='macro')
print("Accuracy score for Decision Tree:",a*100)
print("F1 Score:",b*100)
```

```python
a=accuracy_score(yte,yp2)
b=f1_score(yte,yp2,average='macro')
print("Accuracy score for Naive Bayes Tree:",a*100)
print("F1 Score:",b*100)
```

```python
a=accuracy_score(yte,yp3)
b=f1_score(yte,yp3,average='macro')
print("Accuracy score for K Neighbors classifier:",a*100)
print("F1 Score:",b*100)
```
